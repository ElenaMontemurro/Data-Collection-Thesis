```python
import pandas as pd
from ast import literal_eval
import os
```

```python
rootdir='Cortina\\'

first=1
#consider all restaurants folders
for subdir,dirs,files in os.walk(rootdir):
    if first==1:
        first=0
        continue
    #get restaurant's name
    name=subdir.replace(rootdir,'')
    name=name.rstrip('.')
       
    print(name)
    #load excel file with restaurant's details
    df_raw=pd.read_excel("Cortina\\"+name+"\\"+name+".xlsx")
    df_raw=df_raw.replace('Not present', None)
    #tranform data into a correct format
    columns=df_raw.loc[:,"Unnamed: 0"]
    data=df_raw.loc[:,"Place"]
    data=data.tolist()
    columns = columns.tolist()
    df = pd.DataFrame(data, index= columns).T
    #save in a csv file
    df.to_csv("Cortina\\"+name+"\\"+name+".csv", index=False)
```
