```python
import pandas as pd
from ast import literal_eval
import json
import os
```

```python
rootdir='Firenze\\'

first=1
#for all restaurants
for subdir,dirs,files in os.walk(rootdir):
    if first==1:
        first=0
        continue
    #get restaurant's name
    name=subdir.replace(rootdir,'')
    
    try:    
        print(name)
        #load restaurants reviews excel file
        df_raw=pd.read_excel("Firenze\\"+name+"\\"+name+" reviews.xlsx")
        data_raw = df_raw.loc[:, "collected_data"]
        data_raw = data_raw.apply(literal_eval)
        data_raw = data_raw.tolist()
        df = pd.DataFrame(data_raw)
        #make some preprocessing steps
        df = df.replace(r'\n',' ', regex=True) 
        df['caption'] = df['caption'].str.replace(';',' ') 
        #print(df['name'].count())
        #remove duplicates
        df=df.drop_duplicates(keep='last')
        dup=df[df.duplicated(subset=['user_id','visit_date','date','title'], keep=False)]
        df=df.drop_duplicates(subset=['user_id','visit_date','date', 'title'], keep='last')
        #if len(dup)>0:
            #print(dup)
        #print(df['name'].count())
        #save into a csv file
        df.to_csv("Firenze\\"+name+"\\"+name+" reviews.csv", header=True, index=False)
    except Exception as e:
        print()
        continue
   
```
